```python
import json
```

```python
file = open("_chat.txt","rb")
```

```python
chats = str(file.read())
```

```python
chats = chats.split("\\r\\n")
```

```python
class MSG:
    
    def __init__(self,chat):
        self.raw_text = chat
        
    def createMSG(self):
        temp = self.raw_text.split()
        self.date = temp[0][temp[0].find('[')+1:-1]
        self.time = temp[1][:-1]
        self.sender = temp[2][:-1]
        self.data = " ".join(temp[3:])
        
    def getJSON(self):
        self.JSON = {}
        self.JSON["date"] = self.date
        self.JSON["sender"] = self.sender
        self.JSON["data"] = self.data
        self.JSON["time"] = self.time
        return self.JSON
```

```python
messages=[]
with open("out.json","w") as out:
    for ind,chat in enumerate(chats[:-1]):
        msg = MSG(chat)
        try:
            msg.createMSG()
        except:
            print(ind)
            print("Error at: {} : {}".format(ind,chat))
        messages.append(msg.getJSON())
    json.dump(messages[:1000],out)
print("total msgs dumped: ",len(messages))
```

```python
ll = {}
for ind,msg in enumerate(messages):
    l = len(msg["data"])
    if l in ll:
        ll[l].append(ind)
    else:
        ll[l]=[ind]
for k in ll:
    print(ll[k][0])
```
